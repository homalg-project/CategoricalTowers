In this notebook we *compute* the algorithm computing the braiding isomorphism $\gamma: a \times b \to b \times a$ in a cartesian (monoidal) category, i.e., a category with finite products.

```julia
using CapAndHomalg
```

```julia
LoadPackage( "FiniteCocompletion" )
```

```julia
LoadPackage( "Algebroids" )
```

```julia
LoadPackage( "LazyCategories" );
```

```julia
Q = RightQuiver( "Q(a,b)[]" )
```

```julia
C = FreeCategory( Q )
```

```julia
SetName( C.a, g"C.a" ); SetName( C.b, g"C.b" );
```

```julia
PC = FiniteStrictProductCompletion( C )
```

```julia
Display( PC )
```

```julia
ModelingCategory( PC )
```

```julia
a = PC.a
```

```julia
b = PC.b
```

```julia
ab = DirectProduct( a, b )
```

```julia
Display( ab )
```

```julia
ba = DirectProduct( b, a )
```

```julia
Display( ba )
```

```julia
HomStructure( ab, ba )
```

```julia
hom_ab = MorphismsOfExternalHom( ab, ba )
```

```julia
γ_ab  = hom_ab[1]
```

```julia
Source( γ_ab ) == ab
```

```julia
Range( γ_ab ) == ba
```

```julia
IsWellDefined( γ_ab )
```

```julia
Display( γ_ab )
```

```julia
HomStructure( ba, ab )
```

```julia
hom_ba = MorphismsOfExternalHom( ba, ab )
```

```julia
γ_ba  = hom_ba[1]
```

```julia
Source( γ_ba ) == ba
```

```julia
Range( γ_ba ) == ab
```

```julia
IsWellDefined( γ_ba )
```

```julia
Display( γ_ba )
```

```julia
InverseForMorphisms( γ_ab ) == γ_ba
```

```julia
γ_ab == CartesianBraiding( a, b )
```

```julia
LPC = LazyCategory( PC, primitive_operations = true )
```

```julia
Emb = EmbeddingFunctorOfUnderlyingCategory( LPC )
```

```julia
Display( Emb )
```

```julia
F = PreCompose( CoYonedaEmbeddingOfUnderlyingCategory( PC ), Emb )
```

```julia
Display( F )
```

```julia
G = ExtendFunctorToFiniteStrictProductCompletion( F )
```

```julia
Display( G )
```

```julia
Gγ_ab = G( γ_ab )
```

```julia
IsWellDefined( Gγ_ab )
```

```julia
Visualize( Gγ_ab )
```

```julia
La = LPC.a
```

```julia
Lb = LPC.b
```

```julia
Visualize( CartesianBraiding( La, Lb ) )
```
