## The image of a polynomial map

The following polynomial map
$$
  \varphi:
  \begin{cases} \mathbb{Q}^6 &\to \mathbb{Q}^4,\\
  (a,b,c,d,e,f) &\mapsto (y_1,y_2,y_3,y_4) := (ace+bdf, acf+bde, ade+bcf, bce+adf)
  \end{cases}
$$
considered as Example 5.1 in [HMS18]. It represents the Cavender–Farris–Neyman model, a type of group-based model, which is aspecial Markov process on trees.

[HMS18] Corey Harris, Mateusz Michałek, and Emre Can Sertöz, Computing images of polynomial maps, (arXiv:1801.00827), 2018.


```julia
using CapAndHomalg
```

```julia
LoadPackage( "ZariskiFrames" )
```

```julia
UseSystemSingular(false)
```

```julia
ℚ = HomalgFieldOfRationalsInSingular( )
```

```julia
B = ℚ["y1..4"]
```

```julia
T = ℚ["a,b,c,d,e,f"]
```

```julia
ϕ = RingMap( HomalgMatrix( "[ a*c*e+b*d*f, b*d*e+a*c*f, a*d*e+b*c*f, b*c*e+a*d*f ]", 4, 1, T ),  B, T );
```

```julia
Display( ϕ )
```

```julia
im = ConstructibleImage( ϕ )
```

```julia
FactorizeObject( im )
```

```julia
Display( im )
```

```julia
Visualize( im )
```

```julia
Visualize( im, all = true )
```

```julia
imf = ConstructibleImage( ϕ, freeness = true )
```

```julia
FactorizeObject( imf )
```

```julia
Display( imf )
```

```julia
imf == im
```

```julia
Visualize( imf )
```

```julia
Visualize( imf, all = true )
```
