As we will see below, the law of excluded middle is not valid in $\mathbf{FinBouquets}$. Consequently, the (external) axiom of choice cannot be valid in $\mathbf{FinBouquets}$ either:

```julia
LL = Coproduct( L, L )
```

```julia
L2 = CreateBouquet( 1, [ 0, 0 ] )
```

```julia
ε = CreateBouquetMorphism( LL, [ 0, 0 ], [ 0, 1 ], L2 )
```

```julia
IsWellDefined( ε )
```

```julia
IsEpimorphism( ε )
```

```julia
IsSplitEpimorphism( ε )
```
