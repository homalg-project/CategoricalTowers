$\mathbf{FinBouquets}$ is not a boolean topos:

```julia
⊥ = TruthMorphismOfFalse( FinBouquets )
```

```julia
⊤ = TruthMorphismOfTrue( FinBouquets )
```

LEM is not a tautology in $\mathbf{FinBouquets}$:

```julia
LEM = EmbeddingOfUnionSubobject( ⊤, ⊥ )
```

However, its double negation is a tautology in every topos. To this (for $\mathbf{FinBouquets}$) postcompose the classifying morphism of the LEM (as a subobject of $\Omega$) with the double negation closure operator $\neg\neg$. The resulting composed morphism is the classifying morphism of the full subobject of $\Omega$:

```julia
notnotLEM = SubobjectOfClassifyingMorphism( PreCompose( ClassifyingMorphismOfSubobject( LEM ), notnot ) )
```

```julia
notLEM = EmbeddingOfPseudoComplementSubobject( LEM )
```

```julia
nnLEM = EmbeddingOfPseudoComplementSubobject( notLEM )
```

```julia
nnLEM == notnotLEM
```
