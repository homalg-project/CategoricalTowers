If $F: \text{FreeCategory}(q) \to C$, then $F$ is uniquely determined by its
restriction to the generating quiver $q$.

$$
\text{Hom}_{(\text{Cats})}(\text{FreeCategory}(q),C) \cong \text{Hom}_{(\text{Quivs})}(q,\text{ForgetfulFunctor}(C))
$$

We exploit this in our data structure, a functor
$F:\text{FreeCategory}(q) \to C$ is internally defined on $q$ and
extended to a morphism in $\text{FreeCategory}(q)$ when needed.

More generally, if $B=\langle q \rangle$ is the free category construction in some categorical doctrine,
then functors need only be defined on the generating quiver $q$.

The notion of "generation" depends on the forgetful functor.

Let $D: I \to C$ be a functor. A __cone__
of $D$ consists of the following data:
* An object $S \in C$;
* A dependent function $s$ mapping an object $i \in I$
to a morphism $s(i): S \to D(i)$ with the specification: For all
$i, j \in I, \iota: i \to j$, we have $D(\iota) \circ s(i)=s(j)$.

The __weak limit__ is an algorithm that takes as input a functor
$D: I \to C$ where $I$ is called a diagram category
and outputs the following data:

(1) A cone of $D$ given by the data
$\left(\lim D, \{\lambda(i):\lim D \to D(i)\}_{i\in I}\right)$;

(2) A dependent function $u$ mapping every cone $\tau =
(T,\{\tau(i):T \to D(i)\}_{i\in I})$ to a morphism
$u(\tau):T \to \lim D$ with the specification:
$\lambda(i) \circ u(\tau) = \tau(i)$ for all $i \in I$.

A weak limit is called __limit__, if for any other
dependent function $v$ satisfying (2), we have $u=v$.

First we load the Julia package `CapAndHomalg` giving access to the $\texttt{GAP}$ software projects $\mathtt{QPA2}$, $\mathtt{CAP}$ and $\mathtt{homalg}$:

```julia
using CapAndHomalg
```

Then load the $\mathtt{CAP}$-based package $\mathtt{FunctorCategories}$:

```julia
LoadPackage("FunctorCategories")
```

The field of rationals:

```julia
ℚ = HomalgFieldOfRationals()
```

Define the quiver $q$ with vertices $x,y,z,w$ and arrows $a,b,c,d$:

```julia
q = RightQuiver("q(x,y,z,w)[a:x->y,b:y->z,c:z->w,d:z->w]")
```

$D$ is defined to be the free category generated by the quiver $q$:

```julia
D = FreeCategory(q)
```

Define the quotient category of $E$ by the relation $bc = bd$. Such relations can only express equalities between parallel morphisms:

```julia
E = D / [[D.bc, D.bd]]
```

Construct the $\mathbb Q$-linear closure category $L$ of the finitely presented category $E$:

```julia
L = ℚ[E]
```

Relations in linear categories are generators of a two-sided ideal. We consider the relation $abc=0$:

```julia
B = L / [L.abc]
```

The relations of $B$ are the relations defined for the category $E$ combined with the relations for the linear category $L$:

```julia
UnderlyingQuiverAlgebra(B)
```

Obtain the category $A$ of matrices over $\mathbb Q$ as the range category of the homomorphism structure of $B$:

```julia
A = RangeCategoryOfHomomorphismStructure(B)
```

Construct the category $H$ of functors from $B$ to $A$:

```julia
H = FunctorCategory(B, A)
```

Define the matrices which will be the images of the generating morphisms of $B$ (corresponding to the morphisms of $q$) under a functor in $H$:

```julia
mat_a = [[2,3,1], [1,0,-1]]
```

```julia
mat_b = [[2,0,-1,7], [3,2,1,5], [-2,5,6,-7]]
```

```julia
mat_c = [[-8],[2],[0],[3]] #mat_c = SyzygiesOfColumns(HomalgMatrix(mat_a, ℚ) * HomalgMatrix(mat_b, ℚ))
```

```julia
mat_d = [[-14],[35],[-33],[0]] #mat_d = LeftDivide(HomalgMatrix(mat_b, ℚ), HomalgMatrix(mat_b, ℚ) * HomalgMatrix(mat_c, ℚ))
```

We expect the composition of $\iota$ and $\eta_2 := \eta^2$ to be zero, since the composition of $\iota$ and $\eta$ is zero:

```julia
ι_2 = KernelEmbedding(η_2)
```

We expect the kernel embedding $\iota$ of $\eta$ to be liftable along the dominating kernel embedding $\iota_2$ of $\eta_2 := \eta^2$:

```julia
IsLiftable(ι, ι_2)
```

Now we compute an explicit lift:

```julia
λ = Lift(ι, ι_2)
```

In fact, the lift is unique since $\iota_2$ is a monomorphism (beware of the change of syntax!):

```julia
μ = LiftAlongMonomorphism(ι_2, ι)
```

```julia
λ == μ
```

We do not necessarily expect $\iota_2$ to be liftable along $\iota$:
